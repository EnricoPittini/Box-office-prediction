## FUNZIONI PER LA VALUTAZIONE E SELEZIONE DI MODELLI

```python
#Funzioni che serviranno dopo per le valutazioni dei modelli usando bias/variance/error.


%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from sklearn.utils import resample
from sklearn.preprocessing import MinMaxScaler

# model in realtà è un meta-modello. X e y sono il mio dataset/training set.
def compute_bias_variance_error(model ,X ,y ,N_TESTS = 20):
    
    # Voglio che tale vettore sia una matrice con tante righe quanto N_TESTS (ogni riga è un particolare modello) e tante 
    # colonne quanto ogni punto di X (ogni colonna è un punto del dataset).
    # Riga i ---> ci sono le predizioni fatte dal modello i su tutti i punti del dataset
    # Colonna j ---> ci sono le predizioni fatte sul punto j da tutti gli N_TESTS modelli.
    vector_ypred = []
    
    # Itero su N_TESTS. Ad ogni iterazione creo un nuovo specifico modello i.
    for i in range(N_TESTS):
        # Prendo un campione del dataset.
        Xs, ys = resample(X,y, n_samples=int(0.67*len(y)) )
        
        # Estraggo il modello i dal campione (usando il meta-modello)
        model.fit(Xs,ys)
        
        # Aggiungo le predizioni fatte dal modello i su tutti i punti del dataset.
        vector_ypred.append(list(model.predict(X)))
        
        
    vector_ypred = np.array(vector_ypred) # Trasformo in numpy
    
    # Vettore che ha tanti elementi quanti i punti del dataset e per ciascuno ha il relativo bias calcolato sugli N_TESTS 
    # modelli
    vector_bias = (y - np.mean(vector_ypred, axis=0))**2 # ATTENZIONE
    
    #Vettore che ha tanti elementi quanti i punti del dataset e per ciascuno ha la relativa variance calcolata sugli N_TESTS 
    # modelli
    vector_variance = np.var(vector_ypred, axis=0)
    
    # Vettore che ha tanti elementi quanti i punti del dataset e per ciascuno ha il relativo error calcolato sugli N_TESTS 
    # modelli
    vector_error = np.sum((vector_ypred - y)**2, axis=0)/N_TESTS    
    # Se y array matrice (n,1)
    #vector_error = np.sum((vector_ypred - y.reshape(y.shape[0],1))**2, axis=0)/N_TESTS    
    
    bias = np.mean(vector_bias)
    variance = np.mean(vector_variance)
    error = np.mean(vector_error)
    
    return bias,variance,error
    

def model_selection_BiasVarianceError(model_list, X, y, scale=False ,plotta=False, xvalues=None, xlabel="Complessità"):
    
    if(scale):
        scaler = MinMaxScaler()
        scaler.fit(X)
        X = scaler.transform(X)
    
    biasVarianceError_list = []
    
    for model in model_list:
        biasVarianceError_list.append(list(compute_bias_variance_error(model ,X ,y)))
        
    biasVarianceError_list = np.array(biasVarianceError_list) # to numpy
    
    if(plotta):
        
        if(xvalues is None):
            xvalues = range(len(model_list))
        
        fig, ax = plt.subplots(figsize=(6,6))

        ax.plot(xvalues,biasVarianceError_list[:,0], 'o:', label='Bias$^2$')
        ax.plot(xvalues,biasVarianceError_list[:,1], 'o:', label='Variance')
        ax.plot(xvalues,biasVarianceError_list[:,2], 'o:', label='Error')
        ax.set_xlabel(xlabel)
        ax.grid()
        ax.legend()
        
    return biasVarianceError_list, np.argmin(biasVarianceError_list,axis=0)[2]
```

```python
#Funzioni che serviranno dopo per le valutazioni dei modelli usando bias/variance/error.


from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, accuracy_score
from sklearn.preprocessing import MinMaxScaler

def compute_train_val_test(model ,X_train_80, X_test, y_train_80, y_test, regr=True):
    
    scoring=""
    if regr:
        scoring="neg_mean_squared_error"
    else:
        scoring="accuracy"
    
    # Cross validation
    scores = cross_val_score(model, X_train_80, y_train_80, cv=5, scoring=scoring) ###CV
    val_acc = scores.mean()
    if regr:
        val_acc = -val_acc
    
    model.fit(X_train_80,y_train_80)
    
    train_acc=0
    test_acc=0
    if regr:
        train_acc = mean_squared_error(y_true=y_train_80, y_pred=model.predict(X_train_80))
        test_acc = mean_squared_error(y_true=y_test, y_pred=model.predict(X_test))
    else:
        train_acc = accuracy_score(y_true=y_train_80, y_pred=model.predict(X_train_80))
        test_acc = accuracy_score(y_true=y_test, y_pred=model.predict(X_test))
    
    return train_acc, val_acc, test_acc

def model_selection_TrainValTest(model_list, X, y, regr=True, scale=False, plotta=False, plottaTrain=False, 
                                         plottaTest=False, xvalues=None, xlabel="Complessità"):
    
    X_train_80, X_test, y_train_80, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
    
    if(scale):
        scaler = MinMaxScaler()
        scaler.fit(X_train_80)
        X_train_80 = scaler.transform(X_train_80)
        X_test = scaler.transform(X_test)
    
    trainValTest_list = []
    
    for model in model_list:
        trainValTest_list.append(list(compute_train_val_test(model ,X_train_80, X_test, y_train_80, y_test, regr=regr)))
        
    trainValTest_list = np.array(trainValTest_list) # to numpy
    
    if(plotta):
        
        if(xvalues is None):
            xvalues = range(len(model_list))
        
        fig, ax = plt.subplots(figsize=(6,6))

        if plottaTrain:
            ax.plot(xvalues,trainValTest_list[:,0], 'o:', label='Train')
        ax.plot(xvalues,trainValTest_list[:,1], 'o:', label='Validation')
        if plottaTest:
            ax.plot(xvalues,trainValTest_list[:,2], 'o:', label='Test')
        ax.set_xlabel(xlabel)
        ax.grid()
        ax.legend()
        
    if regr:
        return trainValTest_list, np.argmin(trainValTest_list,axis=0)[1]
    return trainValTest_list, np.argmax(trainValTest_list,axis=0)[1] 
```

# LETTURA E PULIZIA DATASET (NUM,CAT,DATA)

```python
import pandas as pd

def cleaning_data():
    
    data_url = "C:\\Users\\Enrico\\Desktop\\INFORMATICA_TRIENNALE\\dispense informatica\\Terzo anno\\DataWebMining\\Progetto\\train.csv"
    df = pd.read_csv(data_url)
    
    feat_list=[]
    
    # Castiamo a tipo int il tipo della colonna "id"
    df["id"] = df["id"].astype(int)
    # Settiamo la colonna "id" come indice del datframe (Ogni film è indicizzato con il suo id)
    #df.set_index("id",inplace=True)

    # Castiamo a tipo numerico (float) il tipo delle colonne che hanno come valori numeri
    df["budget"] = df["budget"].astype(float)
    #df["revenue"] = df["revenue"].astype(float)
    
    # PULIZIA FEATURES NUMERICHE     
    # Dummy runtime
    df["runtime_miss"] = (df["runtime"].isna() | df["runtime"]<=0).astype(int)  # NON MESSO
    # Al posto dei missing values mettiamo i valori medi.
    df["runtime"].fillna(df["runtime"].mean(),inplace=True) 
    # Runtime non significativo
    df["runtime"] = df["runtime"].map(lambda x : x if x>0 else df["runtime"].mean())
    # Budget non significativo
    df["budget_miss"] = (df["budget"]<=1000).astype(int)
    df["budget"] = df["budget"].map(lambda x : x if x>1000 else df["budget"].mean())   
    # REVENUE ALLA FINE
    feat_list.extend(["budget","budget_miss","popularity","runtime"])
    
    # PULIZIA FEATURES CATEGORIALI
    # belongs_to_collection
    df["belongs_to_collection"] = (~(df["belongs_to_collection"].isna())).astype(int)
    # homepage
    df["homepage"] = (~(df["homepage"].isna())).astype(int)
    # original_language
    """df["original_language"].replace(lang_in_other,"other",inplace=True) # ALT 1
    df = pd.concat([df,pd.get_dummies(df["original_language"])],axis=1)
    df.drop(columns=["original_language"],inplace=True)"""
    df["original_language"] = (df["original_language"]=="en").astype(int) # ALT 2 BEST
    
    feat_list.extend(["belongs_to_collection","homepage","original_language"])
    
    
    # PULIZIA FEATURE DATA
    # year
    df["year"] = df["release_date"].map(lambda s : int(s.split("/")[2]))
    df["year"] = df["year"].map(lambda y : 1900+y if (y>=20 and y<=99) else 2000+y)
    # month
    df["month"] = df["release_date"].map(lambda s : int(s.split("/")[0])-1) # da 0 a 11, per mapping migliore con indici
    
    best_months = np.argsort([df["revenue"][df["month"]==m].mean() for m in range(0,12)])[::-1]
    def transform(m):   # ALT 4
        if m in best_months[0:4]:
            return "month_group_1"
        elif m in list(best_months[4:8]):
            return "month_group_2"
        elif m in list(best_months[8:12]):
            return "month_group_3"
        else:
            raise RuntimeError("Mese invalido: ",m)      
    df["month"].map(transform)
    newdf = pd.get_dummies(df["month"].map(transform))
    for col in newdf.columns:
        df[col] = newdf[col]
    
    feat_list.extend(["year","month_group_1","month_group_2","month_group_3"])
    # REVENUE
    df["revenue"] = df["revenue"].astype(float)
    # Revenue non significativo
    df = df[df["revenue"]>1000]
    
    feat_list.append("revenue")
    
    dataframe = df[feat_list].values
    # Splitto nelle features e nella colonna da prevedere
    n_feat = dataframe.shape[1]
    y = dataframe[:,n_feat-1]#.reshape(df.shape[0],1) # Vettore
    X = dataframe[:,:n_feat-1] # Matrice
    # X e y insieme costituiscono il training set
    
    scaler=MinMaxScaler()
    scaler.fit(y.reshape(df.shape[0],1))
    y = scaler.transform(y.reshape(df.shape[0],1)).reshape(df.shape[0],)
    
    return X,y ,df

X , y, df  = cleaning_data()
print(X.shape)
print(y.shape)
```

```python
# 'Drama', 'Comedy', 'Thriller', 'Action', 'Romance', 'Crime', 'Adventure', 'other_genre'

def add_gen_1(df):
    
    feat_list = []
    
    ## PULIZIA FEATURES INSIEMISTICHE
    
    # Puliamo la colonna "genres". Per ora è una colonna di oggetti dove ogni oggetto è una stringa di lista di dizionari.
    # Togliamo i valori nulli
    df["genres"] = df["genres"].fillna("{}")
    # Puliamo
    # Ogni elemento lo rendiamo lista di dizionari
    df["genres"]=df["genres"].map(lambda obj : list(eval(str(obj))))
    # Ogni lista di dizionari lo rendiamo un set
    df["genres"] = df["genres"].map(lambda lst : {dct['name'] for dct in lst})
    
    generi = set([])
    for film in list(df.index.values):
        generi |= df.loc[film,"genres"]
    generi = list(generi)
    
    # METODO 1
    numFilm_per_genre = [ (df["revenue"][[ gen in s for s in df["genres"] ]]).shape[0] for gen in generi]
    genres_sorted = np.argsort(numFilm_per_genre)[::-1]
    generi_1 = [ generi[gen] for gen in genres_sorted[:9] ] # 7 # 10
    generi_1.append("other_genre")
    generi_in_other = set([ generi[gen] for gen in genres_sorted[9:] ]) # 7 # 10
    for gen in generi_1:
        if gen=="other_genre":
            continue
        #df[gen] = pd.Series([gen in s for s in df["genres"]]).astype(int)
        df[gen] = df["genres"].map(lambda s : gen in s).astype(int)
        #print(df[df[gen].isna()])
    #df["other_genre"] = pd.Series([(generi_in_other&s!=set([])) for s in df["genres"]]).astype(int)
    df["other_genre"] = df["genres"].map(lambda s : generi_in_other&s!=set([])).astype(int)
        
    feat_list.extend(generi_1)
    
    return df[feat_list].values

newX = add_gen_1(df)
X_tmp = np.append(X,newX,axis=1)
X_tmp.shape
```

```python
# 'Drama', 'Comedy', 'Thriller', 'Action', 'Romance', 'Crime', 'Adventure', 'other_genre'

from textblob import TextBlob
from nltk.corpus import stopwords
stopwords=set(stopwords.words('english'))

def lemmatize_with_postag(sentence):
    sent = TextBlob(sentence).lower()
    tag_dict = {"J": 'a', 
                "N": 'n', 
                "V": 'v', 
                "R": 'r'}
    words_and_tags = [(w, tag_dict.get(pos[0], 'n')) for w, pos in sent.tags]    
    lemmatized_list = [wd.lemmatize(tag) for wd, tag in words_and_tags]
    return lemmatized_list

def add_title_1(df):
    
    feat_list = []
    
    df["title"] = df["title"].map(lambda s : set(lemmatize_with_postag(s)) - stopwords)
    
    words = set([])
    for film in list(df.index.values):
        words |= df.loc[film,"title"]
    words = np.array(list(words))
    
    numFilm_per_wordTitle = np.array([ (df["revenue"][[ c in s for s in df["title"] ]]).shape[0] for c in words])
    best_wordTitle_num = np.argsort(numFilm_per_wordTitle)[::-1]
    
    selected_words = words[best_wordTitle_num[3:10]]
    for w in selected_words:
        df[w+"_title"] = df["title"].map(lambda s : w in s).astype(int)
    
    feat_list.extend([w+"_title" for w in selected_words])
    
    return df[feat_list].values

newX = add_title_1(df)
X_tmp = np.append(X,newX,axis=1)
X_tmp.shape
```
